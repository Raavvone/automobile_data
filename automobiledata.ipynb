```python
import pandas as pd
import numpy as np
```

```python
file_="/content/automobile.csv"
```

```python
data_=pd.read_csv(file_)
```

```python
data_
```

```python
data_.info()
```

```python
data_.columns
```

```python
len(data_.columns)
```

```python
data_.replace("?", np.nan, inplace=True)
```

```python
data_
```

```python
data_["normalized-losses"].astype(float).mean()
```

```python
data_.dropna(subset=["price"], axis=0, inplace=True)
```

```python
data_
```

```python
data_["normalized-losses"].fillna(data_["normalized-losses"].astype(float).mean(), inplace=True)
```

```python
# Convert data types from object to float
data_[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']] = data_[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']].astype(float)
```

```python
data_.dtypes
```

```python
data_.drop_duplicates(inplace=True)
```
